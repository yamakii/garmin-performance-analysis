````python
import sys
from datetime import datetime, timedelta
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import polars as pl

# ãƒ—ãƒ­ã‚¸ã‚§ã‚¯ãƒˆãƒ«ãƒ¼ãƒˆã‚’PATHã«è¿½åŠ 
project_root = Path().absolute().parent.parent
sys.path.insert(0, str(project_root))

from tools.utils.llm_safe_data import safe_json_output, safe_load_export
````

## ã‚¹ãƒ†ãƒƒãƒ—1: ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒ«ã®å–å¾—ï¼ˆã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆï¼‰

å®Ÿéš›ã®ç’°å¢ƒã§ã¯ã€ä»¥ä¸‹ã®ã‚ˆã†ã«å„ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ã®ãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒ«ã‚’å–å¾—ã—ã¾ã™ï¼š

```python
profiles = []
for activity_id in [12345, 12346, 12347, 12348, 12349]:
    profile = mcp__garmin_db__profile(
        table_or_query=f"SELECT * FROM splits WHERE activity_id = {activity_id}"
    )
    profiles.append(profile)
```

````python
# ãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒ«ã®ãƒ¢ãƒƒã‚¯ï¼ˆ5ã¤ã®ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ï¼‰
base_date = datetime(2025, 1, 1)
activity_ids = [12345, 12346, 12347, 12348, 12349]

profiles = []
for i, activity_id in enumerate(activity_ids):
    profile = {
        'activity_id': activity_id,
        'date': (base_date + timedelta(weeks=i)).strftime('%Y-%m-%d'),
        'distance_km': 10.0 + np.random.uniform(-0.2, 0.2),
        'duration_seconds': 2700 + i * 60,  # æ”¹å–„å‚¾å‘ï¼ˆé€Ÿããªã£ã¦ã„ã‚‹ï¼‰
        'avg_pace_sec_per_km': 270 + i * 6,  # æ”¹å–„å‚¾å‘ï¼ˆé…ããªã£ã¦ã„ã‚‹ â†’ ç–²åŠ´ï¼Ÿï¼‰
        'avg_heart_rate': 150 - i * 2,  # æ”¹å–„å‚¾å‘ï¼ˆå¿ƒæ‹ãŒä¸‹ãŒã£ã¦ã„ã‚‹ â†’ åŠ¹ç‡åŒ–ï¼‰
        'hr_zone_distribution': {
            'zone_1': 5 + i * 2,
            'zone_2': 15 + i * 3,
            'zone_3': 40 - i * 2,
            'zone_4': 30 - i * 3,
            'zone_5': 10
        }
    }
    profiles.append(profile)

print("ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ãƒ—ãƒ­ãƒ•ã‚¡ã‚¤ãƒ«:")
for profile in profiles:
    print(f"\n  Activity {profile['activity_id']} ({profile['date']}):")
    print(f"    è·é›¢: {profile['distance_km']:.2f}km")
    print(f"    æ™‚é–“: {profile['duration_seconds']//60}:{profile['duration_seconds']%60:02d}")
    print(f"    å¹³å‡ãƒšãƒ¼ã‚¹: {profile['avg_pace_sec_per_km']//60}:{profile['avg_pace_sec_per_km']%60:02d}/km")
    print(f"    å¹³å‡å¿ƒæ‹: {profile['avg_heart_rate']:.0f} bpm")

print("\nãƒˆãƒ¼ã‚¯ãƒ³ã‚³ã‚¹ãƒˆ: ç´„625ãƒˆãƒ¼ã‚¯ãƒ³ï¼ˆ5ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ Ã— 125ãƒˆãƒ¼ã‚¯ãƒ³ï¼‰")
````

## ã‚¹ãƒ†ãƒƒãƒ—2: å¿ƒæ‹ãƒ‡ãƒ¼ã‚¿ã®ã‚¨ã‚¯ã‚¹ãƒãƒ¼ãƒˆï¼ˆã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆï¼‰

å®Ÿéš›ã®ç’°å¢ƒã§ã¯ã€å„ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ã®å¿ƒæ‹ãƒ‡ãƒ¼ã‚¿ã‚’ã‚¨ã‚¯ã‚¹ãƒãƒ¼ãƒˆã—ã¾ã™ï¼š

```python
export_results = []
for activity_id in activity_ids:
    export_result = mcp__garmin_db__export(
        query=f"""
            SELECT timestamp, heart_rate
            FROM time_series_metrics
            WHERE activity_id = {activity_id}
        """,
        format="parquet"
    )
    export_results.append(export_result)
```

````python
# ãƒ¢ãƒƒã‚¯ãƒ‡ãƒ¼ã‚¿: å„ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ã®å¿ƒæ‹ãƒ‡ãƒ¼ã‚¿
np.random.seed(42)
export_results = []

for i, (activity_id, profile) in enumerate(zip(activity_ids, profiles, strict=False)):
    duration = profile['duration_seconds']
    timestamps = np.arange(0, duration)

    # å¿ƒæ‹ã®æ™‚ç³»åˆ—: ã‚¦ã‚©ãƒ¼ãƒ ã‚¢ãƒƒãƒ— â†’ å®‰å®š â†’ ä¸Šæ˜‡ï¼ˆç–²åŠ´ï¼‰
    warmup = np.linspace(120, 150, duration // 10)  # æœ€åˆã®10%ã§ã‚¦ã‚©ãƒ¼ãƒ ã‚¢ãƒƒãƒ—
    steady = np.full(duration * 7 // 10, 150 - i * 2) + np.random.normal(0, 3, duration * 7 // 10)
    fatigue = np.linspace(150 - i * 2, 160 - i * 2, duration - len(warmup) - len(steady))  # æœ€å¾Œã®20%ã§ç–²åŠ´

    heart_rate = np.concatenate([warmup, steady, fatigue])

    df = pl.DataFrame({
        'timestamp': timestamps,
        'heart_rate': heart_rate
    })

    # Parquetã¨ã—ã¦ä¿å­˜
    export_path = Path(f'/tmp/export_activity_{activity_id}.parquet')
    df.write_parquet(export_path)

    export_result = {
        'activity_id': activity_id,
        'handle': str(export_path),
        'rows': len(df),
        'size_mb': export_path.stat().st_size / (1024 * 1024),
        'columns': ['timestamp', 'heart_rate']
    }
    export_results.append(export_result)

print("ã‚¨ã‚¯ã‚¹ãƒãƒ¼ãƒˆçµæœ:")
for result in export_results:
    print(f"  Activity {result['activity_id']}: {result['rows']} è¡Œ")

print("\nãƒˆãƒ¼ã‚¯ãƒ³ã‚³ã‚¹ãƒˆ: ç´„125ãƒˆãƒ¼ã‚¯ãƒ³ï¼ˆ5ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ Ã— 25ãƒˆãƒ¼ã‚¯ãƒ³ï¼‰")
````

## ã‚¹ãƒ†ãƒƒãƒ—3: ãƒ‡ãƒ¼ã‚¿ã‚’ãƒ­ãƒ¼ãƒ‰ã—ã¦æ™‚é–“è»¸ã‚’æ­£è¦åŒ–

å„ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ã®æ™‚é–“è»¸ã‚’0-100%ã«æ­£è¦åŒ–ã—ã¦æ¯”è¼ƒå¯èƒ½ã«ã—ã¾ã™ã€‚

````python
# ãƒ‡ãƒ¼ã‚¿ã‚’ãƒ­ãƒ¼ãƒ‰ã—ã¦æ­£è¦åŒ–
activities_data = []

for export_result, profile in zip(export_results, profiles, strict=False):
    df = safe_load_export(export_result['handle'], max_rows=10000)

    # æ™‚é–“è»¸ã‚’0-100%ã«æ­£è¦åŒ–
    df = df.with_columns([
        (pl.col('timestamp') / pl.col('timestamp').max() * 100).alias('progress_pct')
    ])

    activities_data.append({
        'activity_id': export_result['activity_id'],
        'date': profile['date'],
        'df': df
    })

print(f"âœ“ {len(activities_data)} ã‚¢ã‚¯ãƒ†ã‚£ãƒ“ãƒ†ã‚£ã®ãƒ‡ãƒ¼ã‚¿ã‚’ãƒ­ãƒ¼ãƒ‰ãƒ»æ­£è¦åŒ–")
````

## ã‚¹ãƒ†ãƒƒãƒ—4: é‡ã­åˆã‚ã›ã‚°ãƒ©ãƒ•ã®ç”Ÿæˆ

````python
# é‡ã­åˆã‚ã›ã‚°ãƒ©ãƒ•
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# å·¦: æ­£è¦åŒ–æ™‚é–“è»¸ã§ã®å¿ƒæ‹æ¨ç§»
colors = plt.cm.viridis(np.linspace(0, 1, len(activities_data)))
for i, activity in enumerate(activities_data):
    df_pandas = activity['df'].to_pandas()
    ax1.plot(df_pandas['progress_pct'], df_pandas['heart_rate'],
             label=f"{activity['date']}", color=colors[i], alpha=0.7, linewidth=2)

ax1.set_xlabel('é€²æ— (%)')
ax1.set_ylabel('å¿ƒæ‹ (bpm)')
ax1.set_title('å¿ƒæ‹ã‚¾ãƒ¼ãƒ³æ¨ç§»ã®æ¯”è¼ƒï¼ˆæ­£è¦åŒ–æ™‚é–“è»¸ï¼‰')
ax1.legend()
ax1.grid(True, alpha=0.3)

# å³: å¹³å‡å¿ƒæ‹ã®æ¨ç§»ï¼ˆãƒˆãƒ¬ãƒ³ãƒ‰ï¼‰
avg_hrs = [profile['avg_heart_rate'] for profile in profiles]
dates = [profile['date'] for profile in profiles]

ax2.plot(range(len(avg_hrs)), avg_hrs, 'o-', color='blue', linewidth=2, markersize=8)
ax2.set_xticks(range(len(dates)))
ax2.set_xticklabels(dates, rotation=45, ha='right')
ax2.set_ylabel('å¹³å‡å¿ƒæ‹ (bpm)')
ax2.set_title('å¹³å‡å¿ƒæ‹ã®ãƒˆãƒ¬ãƒ³ãƒ‰')
ax2.grid(True, alpha=0.3)

# ãƒˆãƒ¬ãƒ³ãƒ‰ãƒ©ã‚¤ãƒ³
z = np.polyfit(range(len(avg_hrs)), avg_hrs, 1)
p = np.poly1d(z)
ax2.plot(range(len(avg_hrs)), p(range(len(avg_hrs))), "r--", alpha=0.7,
         label=f'ãƒˆãƒ¬ãƒ³ãƒ‰ (å‚¾ã: {z[0]:.1f} bpm/é€±)')
ax2.legend()

plt.tight_layout()
plot_path = Path('/tmp/multi_activity_comparison_plot.png')
plt.savefig(plot_path, dpi=100, bbox_inches='tight')
plt.show()

print(f"âœ“ ã‚°ãƒ©ãƒ•ã‚’ä¿å­˜: {plot_path}")
````

## ã‚¹ãƒ†ãƒƒãƒ—5: çµ±è¨ˆã‚µãƒãƒªãƒ¼ã®è¨ˆç®—

````python
# çµ±è¨ˆã‚µãƒãƒªãƒ¼
summary = {
    'num_activities': len(profiles),
    'date_range': {
        'start': profiles[0]['date'],
        'end': profiles[-1]['date']
    },
    'performance_trend': {
        'avg_heart_rate': {
            'first': float(profiles[0]['avg_heart_rate']),
            'last': float(profiles[-1]['avg_heart_rate']),
            'change': float(profiles[-1]['avg_heart_rate'] - profiles[0]['avg_heart_rate']),
            'change_pct': float((profiles[-1]['avg_heart_rate'] - profiles[0]['avg_heart_rate']) / profiles[0]['avg_heart_rate'] * 100)
        },
        'avg_pace': {
            'first': float(profiles[0]['avg_pace_sec_per_km']),
            'last': float(profiles[-1]['avg_pace_sec_per_km']),
            'change': float(profiles[-1]['avg_pace_sec_per_km'] - profiles[0]['avg_pace_sec_per_km']),
            'change_pct': float((profiles[-1]['avg_pace_sec_per_km'] - profiles[0]['avg_pace_sec_per_km']) / profiles[0]['avg_pace_sec_per_km'] * 100)
        }
    },
    'hr_zone_trend': {
        'zone_3_change': profiles[-1]['hr_zone_distribution']['zone_3'] - profiles[0]['hr_zone_distribution']['zone_3'],
        'zone_4_change': profiles[-1]['hr_zone_distribution']['zone_4'] - profiles[0]['hr_zone_distribution']['zone_4']
    },
    'plot_path': str(plot_path)
}

# è¦ç´„JSONã‚’å‡ºåŠ›ï¼ˆ1KBåˆ¶é™ï¼‰
summary_json = safe_json_output(summary)
print("è¦ç´„ãƒ‡ãƒ¼ã‚¿ï¼ˆLLMã«è¿”å´ï¼‰:")
print(summary_json)
print(f"\nJSON ã‚µã‚¤ã‚º: {len(summary_json.encode('utf-8'))} ãƒã‚¤ãƒˆ")
print("ãƒˆãƒ¼ã‚¯ãƒ³ã‚³ã‚¹ãƒˆ: ç´„150ãƒˆãƒ¼ã‚¯ãƒ³")
````

## ã‚¹ãƒ†ãƒƒãƒ—6: LLMã«ã‚ˆã‚‹ãƒ‘ãƒ•ã‚©ãƒ¼ãƒãƒ³ã‚¹å¤‰åŒ–ã®è§£èª¬ï¼ˆã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ãƒˆï¼‰

````python
# LLMã®è§£é‡ˆä¾‹
interpretation = f"""
**éå»5å›ã®10kmãƒ©ãƒ³ã®æ¯”è¼ƒåˆ†æ:**

**æœŸé–“**: {summary['date_range']['start']} ~ {summary['date_range']['end']} (5é€±é–“)

**ãƒ‘ãƒ•ã‚©ãƒ¼ãƒãƒ³ã‚¹ãƒˆãƒ¬ãƒ³ãƒ‰:**

1. **å¿ƒæ‹åŠ¹ç‡ã®æ”¹å–„ï¼ˆãƒã‚¸ãƒ†ã‚£ãƒ–ï¼‰**
   - å¹³å‡å¿ƒæ‹: {summary['performance_trend']['avg_heart_rate']['first']:.0f} â†’ {summary['performance_trend']['avg_heart_rate']['last']:.0f} bpm ({summary['performance_trend']['avg_heart_rate']['change']:.0f} bpmã€{summary['performance_trend']['avg_heart_rate']['change_pct']:.1f}%æ¸›å°‘)
   - åŒã˜ãƒšãƒ¼ã‚¹ã§ã‚ˆã‚Šä½ã„å¿ƒæ‹ã§èµ°ã‚Œã‚‹ã‚ˆã†ã«ãªã£ã¦ã„ã¾ã™
   - ã“ã‚Œã¯æœ‰é…¸ç´ èƒ½åŠ›ã®å‘ä¸Šã‚’ç¤ºã—ã¦ã„ã¾ã™ âœ…

2. **ãƒšãƒ¼ã‚¹ã®å¤‰åŒ–ï¼ˆè¦æ³¨æ„ï¼‰**
   - å¹³å‡ãƒšãƒ¼ã‚¹: {summary['performance_trend']['avg_pace']['first']//60:.0f}:{summary['performance_trend']['avg_pace']['first']%60:02.0f} â†’ {summary['performance_trend']['avg_pace']['last']//60:.0f}:{summary['performance_trend']['avg_pace']['last']%60:02.0f} /km ({summary['performance_trend']['avg_pace']['change']:.0f}ç§’ã€{summary['performance_trend']['avg_pace']['change_pct']:.1f}%å¢—åŠ )
   - ãƒšãƒ¼ã‚¹ãŒã‚„ã‚„é…ããªã£ã¦ã„ã¾ã™ãŒã€å¿ƒæ‹ãŒä¸‹ãŒã£ã¦ã„ã‚‹ãŸã‚ã€æ„å›³çš„ãªã‚¤ãƒ¼ã‚¸ãƒ¼ãƒ©ãƒ³ã®å¯èƒ½æ€§ãŒã‚ã‚Šã¾ã™
   - ã¾ãŸã¯ç–²åŠ´ã®è“„ç©ã®å…†å€™ã‹ã‚‚ã—ã‚Œã¾ã›ã‚“ âš ï¸

3. **å¿ƒæ‹ã‚¾ãƒ¼ãƒ³åˆ†å¸ƒã®å¤‰åŒ–**
   - Zone 3ã®æ™‚é–“: {summary['hr_zone_trend']['zone_3_change']:+.0f}%
   - Zone 4ã®æ™‚é–“: {summary['hr_zone_trend']['zone_4_change']:+.0f}%
   - ã‚ˆã‚Šä½ã„å¿ƒæ‹ã‚¾ãƒ¼ãƒ³ã§ã®ãƒˆãƒ¬ãƒ¼ãƒ‹ãƒ³ã‚°ã«ã‚·ãƒ•ãƒˆã—ã¦ã„ã¾ã™
   - ã“ã‚Œã¯æœ‰é…¸ç´ ãƒ™ãƒ¼ã‚¹ã®å¼·åŒ–ã«é©ã—ãŸå¤‰åŒ–ã§ã™ âœ…

**æ¨å¥¨:**
- å¿ƒæ‹åŠ¹ç‡ã®æ”¹å–„ã¯éå¸¸ã«ãƒã‚¸ãƒ†ã‚£ãƒ–ãªã‚µã‚¤ãƒ³ã§ã™
- ãƒšãƒ¼ã‚¹ã®ä½ä¸‹ãŒæ„å›³çš„ãªã‚‚ã®ã‹ç¢ºèªã—ã¦ãã ã•ã„
- ç–²åŠ´ãŒè“„ç©ã—ã¦ã„ã‚‹å ´åˆã¯ã€ãƒªã‚«ãƒãƒªãƒ¼é€±ã‚’è¨­ã‘ã‚‹ã“ã¨ã‚’æ¤œè¨ã—ã¦ãã ã•ã„
- æ¬¡å›ã®ãƒ†ã‚¹ãƒˆã§ã¯ã€åŒã˜ãƒšãƒ¼ã‚¹ã§ã®å¿ƒæ‹å¤‰åŒ–ã‚’è¿½è·¡ã™ã‚‹ã“ã¨ã‚’ãŠå‹§ã‚ã—ã¾ã™
"""

print(interpretation)
````
